```python
import pandas as pd
import os
import kagglehub
import shutil
```

```python
os.makedirs('../data/processed', exist_ok=True)
os.makedirs('../data/datasets', exist_ok=True)
os.makedirs('../data/datasets/flavors_of_cacao', exist_ok=True)
```

```python
# Download da versão mais recente do dataset
path = kagglehub.dataset_download("christopherzaman/flavors-of-cacao-2025")
print("Path temporário do dataset:", path)

# Define o destino relativo ao notebook
dest_dir = os.path.join(os.path.dirname(os.getcwd()), "data", "datasets",'flavors_of_cacao')

# Copia os arquivos do cache do kagglehub para o destino
for arquivo in os.listdir(path):
    src = os.path.join(path, arquivo)
    dst = os.path.join(dest_dir, arquivo)
    shutil.copy2(src, dst)
    print(f"Copiado: {arquivo}")

print(f"\nDataset salvo em: {dest_dir}")
```

```python
df_2 = df.copy()
```

```python
df_2['Rating'] = df_2['Rating'].astype(str)
```

```python
df_2['Rating']
```

```python
df_2['Rating'] = df_2['Rating'].str.replace('.',',')
```

```python
df_2['Cocoa Percent'] = df_2['Cocoa Percent'].str.replace('%','')
```

```python
df_2['Company Location'] = df_2['Company Location'].str.replace('.','')
```

```python
df_2.head()
```

```python
lista = df_2['Company Location'].unique()
lista
```

```python
import pandas as pd
from deep_translator import GoogleTranslator

# 1. Inicializar o tradutor (mais robusto que o googletrans)
translator = GoogleTranslator(source='en', target='pt')

# 2. Pegar os valores únicos para economizar tempo/API
valores_unicos = df_2['Company Location'].unique()

# 3. Criar o dicionário de tradução
mapa_traducao = {}
print("Iniciando tradução...")

for texto in valores_unicos:
    try:
        # A deep-translator não costuma dar o erro de 'NoneType'
        traducao = translator.translate(texto)
        mapa_traducao[texto] = traducao
    except Exception as e:
        print(f"Erro ao traduzir '{texto}': {e}")
        mapa_traducao[texto] = texto

# 4. Aplicar na coluna
df_2['Company Location PT'] = df_2['Company Location'].map(mapa_traducao)

print("Tradução concluída!")
df_2[['Company Location', 'Company Location PT']].head()
```
